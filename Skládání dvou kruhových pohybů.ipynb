# Skládání dvou kruhových pohybů

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T = 2 * np.pi / omega_k  # perioda
t = np.linspace(0, T, 1000)
r_k = 1
omega_k = 1
r_h = 0.5
omega_h = -1
x = r_k * np.cos(omega_k * t) + r_h * np.cos(omega_h * t)
y = r_k * np.sin(omega_k * t) + r_h * np.sin(omega_h * t)
```

```python
plt.plot(x,y)
plt.axis('equal')
plt.title('Skládání dvou kruhových pohybů')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

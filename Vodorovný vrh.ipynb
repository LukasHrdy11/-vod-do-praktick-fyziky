# Vodorovný vrh

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h = 10  # výška
g = 9.81  # tíhové zrychlení
dt = 0.01  # časový krok
t = []
x = []
y = []
t = np.append(t, 0)
x = np.append(x, 0)
y = np.append(y, h)
v_0 = 5  # počáteční rychlost
```

```python
i = 0
while y[i] >= 0:
    x = np.append(x, v_0*t[i])  # vodorovný vrh
    y = np.append(y, h - 0.5*g*t[i]**2)  # svislý vrh
    t = np.append(t, t[i] + dt)
    i += 1
```

```python
fig,ax = plt.subplots()
plt.plot(x, y)
plt.title('Trajektorie vrhu')
plt.xlabel('x [m]')
plt.ylabel('y [m]')
plt.grid()
plt.xlim(0)
plt.ylim(0)
plt.show()
```

Spočítejem vzdálenost vrhu a délku trajektorie

```python
delka_vrhu = x[-1]
print(f'Délka vrhu je {delka_vrhu:.2f} m')
```

```python
def numerical_integration(rychlost, dt):
    N = len(rychlost)
    s = np.zeros(N)
    for i in range(0, N-1):
        s[i+1] = s[i] + rychlost[i] * dt
    return s
```

```python
velocity = np.sqrt(v_0**2 + (g*t)**2)
delka_vrhu_numerical = numerical_integration(velocity, dt)[-1]
print(f'Délka vrhu (numerická integrace) je {delka_vrhu_numerical:.2f} m')
```

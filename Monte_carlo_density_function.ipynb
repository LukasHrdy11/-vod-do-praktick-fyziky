```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
n = 1000000
nbins = 100
data = np.random.random(n)
data
```

```python
plt.hist(data, bins = nbins, density = True)
plt.plot([0,1],[1,1])
plt.show()
```

```python
def monte_carlo_number_generation(a, c, modulo):
    I_current = time.time_ns() 
    for _ in range(100):
        I_next = (a * I_current + c) % modulo
        I_current = I_next 
        
        normalized_result = I_current / modulo
    return normalized_result
```

```python
monte_carlo_number_generation(5, 4, 2**31)
```

```python
hist_data = [monte_carlo_number_generation(5, 4, 2**31) for _ in range(n)]
hist_data
```

```python
plt.hist(hist_data, bins = nbins, density = True)
plt.plot([0,1],[1,1])
plt.show()
```

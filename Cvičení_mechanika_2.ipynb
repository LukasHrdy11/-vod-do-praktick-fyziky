Nakreslení kruhové trajektorie trajektorie o poloměru r 

Parametry pohybu

```python
r = 1  # poloměr kruhové trajektorie
T = 1
v_r = 1  # rychlost zmenšování r
omega = 2 * np.pi / T  # úhlová rychlost   
dt = 0.0001
t = np.arange(0, T+dt, dt)  # časový vektor arange dělá pole a není tam počet, ale krok dt
x = r * np.cos(omega * t)  # x souřadnice
y = r * np.sin(omega * t)  # y souřadnice
v_x = -r*omega*np.sin(omega*t)  # derivace x podle t
v_y = r*omega*np.cos(omega*t)   # derivace y podle t
v_velikost = np.sqrt(v_x**2 + v_y**2) # velikost rychlosti
x_2 = (r - v_r*t) * np.cos(omega * t)  # x souřadnice snižujícím se r
y_2 = (r - v_r*t) * np.sin(omega * t)  # y souřadnice snižujícím se r
v_x2 = - (r - v_r*t) * omega * np.sin(omega*t) - v_r * np.cos(omega*t)  # derivace x podle t se snižujícím se r
v_y2 = (r - v_r*t) * omega * np.cos(omega*t) - v_r * np.sin(omega*t)   # derivace y podle t se snižujícím se r
v_velikost2 = np.sqrt(v_x2**2 + v_y2**2) # velikost rychlosti se snižujícím se r
```

Numerická derivace pro výpočet rychlosti z polohy

```python
import numpy as np

T = 1
dt = 0.0001
r = 1  # poloměr kruhové trajektorie
t = np.arange(0, T + dt, dt)  # časový vektor
omega = 2 * np.pi / T  # úhlová rychlost   

def x_souradnice(time):
    return r * np.cos(omega * time)

def y_souradnice(time):
    return r * np.sin(omega * time)

def numerical_derivation(function, time, time_step):
    deriv = np.zeros(len(time))  # výstupní pole stejné délky
    for i in range(len(time) - 1):
        deriv[i] = (function(time[i+1]) - function(time[i])) / time_step
    deriv[-1] = deriv[-2]  # poslední bod aproximujeme
    return deriv

# výpočet rychlostí v obou směrech
vx = numerical_derivation(x_souradnice, t, dt)
vy = numerical_derivation(y_souradnice, t, dt)

# výsledná velikost rychlosti
v = np.sqrt(vx**2 + vy**2)
```

```python
fig,ax = plt.subplots()
ax.plot(t,y)
ax.plot(t,x)
ax.plot(t,vx)
ax.plot(t,vy)
ax.plot(t,v)
ax.set_title('x a y souřadnice v čase')
ax.set_xlabel('t [s]')
ax.set_ylabel('y a x [m]')
ax.legend(['y souřadnice','x souřadnice','y rychlost','x rychlost','rychlost'])
ax.grid()
plt.show()
```

Vykreslení kmitavého krouživého pohybu 

```python
A = r  # amplituda kmitání
t = np.arange(0, T+dt, dt)  # časový vektor arange dělá pole a není tam počet, ale krok dt
x_3 = (r - A*np.sin(omega*10 * t)) * np.cos(omega * t)  # x souřadnice
y_3 = (r - A*np.sin(omega*10 * t)) * np.sin(omega * t)  # y souřadnice
```

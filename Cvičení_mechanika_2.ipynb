```python
import numpy as np
import matplotlib.pyplot as plt
```

Nakreslení kruhové trajektorie trajektorie o poloměru r 

Parametry pohybu

```python
r = 1  # poloměr kruhové trajektorie
T = 1
omega = 2 * np.pi / T  # úhlová rychlost   
dt = 0.0001
t = np.arange(0, T+dt, dt)  # časový vektor arange dělá pole a není tam počet, ale krok dt
x = r * np.cos(omega * t)  # x souřadnice
y = r * np.sin(omega * t)  # y souřadnice
```

```python
# fig,ax = plt.subplots(figsize(5,5)) můžeme takto zapisovat 
plt.plot(x, y)
plt.axis('equal') # stejné měřítko na osách
plt.title('Trajektorie na kružnici')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
fig,ax = plt.subplots()
ax.plot(t,y)
ax.plot(t,x)
ax.set_title('x a y souřadnice v čase')
ax.set_xlabel('t [s]')
ax.set_ylabel('y a x [m]')
ax.legend(['y souřadnice','x souřadnice'])
ax.grid()
plt.show()
```

Rychlost jako derivace souřadnic

```python
v_x = -r*omega*np.sin(omega*t)  # derivace x podle t
v_y = r*omega*np.cos(omega*t)   # derivace y podle t
```

```python
plt.plot(t,v_x)
plt.plot(t,v_y)
plt.title('Rychlost v čase')
plt.xlabel('t [s]')
plt.ylabel('v_x a v_y [m/s]')
plt.legend(['v_x','v_y'])
plt.grid()
plt.show()
```

Velikost rychlosti

```python
v_velikost = np.sqrt(v_x**2 + v_y**2)
```

```python
plt.plot(t,v_velikost)
plt.title('Velikost rychlosti v čase')
plt.xlabel('t [s]')
plt.ylabel('v [m/s]')
plt.grid()
plt.show()
```

Trajektorii se snižujícím se r

```python
r = 1  # poloměr kruhové trajektorie
T = 1
v_r = 1  # rychlost zmenšování r
omega = 2 * np.pi / T  # úhlová rychlost   
dt = 0.01
t = np.arange(0, T+dt, dt)  # časový vektor arange dělá pole a není tam počet, ale krok dt
x_2 = (r - v_r*t) * np.cos(omega * t)  # x souřadnice
y_2 = (r - v_r*t) * np.sin(omega * t)  # y souřadnice
```

```python
# fig,ax = plt.subplots(figsize(5,5)) můžeme takto zapisovat 
plt.plot(x_2, y_2)
plt.axis('equal') # stejné měřítko na osách
plt.title('Trajektorie na kružnici')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

Vykreslení v závislosti na čase

```python
plt.plot(t, x_2) 
plt.plot(t, y_2)
```

Vykreslení časové závisloti rychlosti na čase

```python
v_x2 = - (r - v_r*t) * omega * np.sin(omega*t) - v_r * np.cos(omega*t)  # derivace x podle t
v_y2 = (r - v_r*t) * omega * np.cos(omega*t) - v_r * np.sin(omega*t)   # derivace y podle t
v_velikost2 = np.sqrt(v_x2**2 + v_y2**2)
```

```python
plt.plot(t, v_x2)
plt.plot(t, v_y2)
plt.title('Rychlost v čase')
```

```python
plt.plot(t, v_velikost2)
plt.title('Velikost rychlosti v čase')
plt.xlabel('t [s]')
plt.ylabel('v [m/s]')
plt.grid()
plt.show()
```

Výpočet dráhy pomocí přibližného integrálu

```python
s = 0
for i in range(len(t)):
    ds = v_velikost2[i] * dt  # přibližný integrál
    s += ds

print('dráha s =', s)  
```

Vykreslení kmitavého krouživého pohybu 

```python
r = 1  # poloměr kruhové trajektorie
T = 1
omega = 2 * np.pi / T  # úhlová rychlost   
dt = 0.01
A = 5  # amplituda kmitání
t = np.arange(0, T+dt, dt)  # časový vektor arange dělá pole a není tam počet, ale krok dt
x_3 = (r - A*np.sin(omega * t)) * np.cos(omega * t)  # x souřadnice
y_3 = (r - A*np.sin(omega * t)) * np.sin(omega * t)  # y souřadnice
```

```python
plt.plot(x_3, y_3)
plt.axis('equal') # stejné měřítko na osách
plt.title('Trajektorie na kružnici s kmitáním')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```
